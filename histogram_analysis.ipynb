```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset from the selected Excel file by specifying the header row (index 3 for the 4th row)
df = pd.read_excel('/content/Population data set.xls', header=3)

# Remove missing values
df = df.dropna()

# Print columns to identify the correct population data column
print(df.columns)

# Select the relevant column for population data, using '2023'
population_data = df["2023"]

# Plot histogram with KDE
plt.figure(figsize=(8,5))
sns.histplot(population_data, bins=20, kde=True)

plt.title("Population Distribution for 2023")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.show()
```
